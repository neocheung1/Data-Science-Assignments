```python
import numpy as np  # useful for many scientific computing in Python
import pandas as pd # primary data structure library
```

```python
df_survey = pd.read_csv('https://cocl.us/datascience_survey_data')

print('Dataset downloaded and read into a pandas dataframe!')
```

```python
#df_survey.head()
df_survey.sort_values(['Very interested'], ascending=False, axis=0, inplace=True)
df_survey.rename(columns={'Unnamed: 0':'Topic'}, inplace=True )
df_survey.set_index('Topic', inplace = True)
df_survey
```

```python
# use the inline backend to generate the plots within the browser
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot') # optional: for ggplot-like style

# check for latest version of Matplotlib
print ('Matplotlib version: ', mpl.__version__) # >= 2.0.0
```

```python
ax = df_survey.plot(kind='bar', figsize=(20, 8),color = ['#5cb85c','#5bc0de','#d9534f'],
                width=.8,fontsize=14)
plt.xlabel('Topic',fontsize=14) # add to x-label to the plot
plt.ylabel('Survey Results',fontsize=14) # add y-label to the plot
plt.legend(labels=df_survey.columns,fontsize=14)
plt.title('Data Science Survey',fontsize=16) # add title to the plot


plt.xticks(fontsize=14)
for spine in plt.gca().spines.values():
    spine.set_visible(False)
plt.yticks([])

# Add this loop to add the annotations
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy() 
    ax.annotate('{:.0%}'.format(height), (x, y + height + 0.01))


plt.show()
```
